```python
#プログラム1｜ライブラリの設定
import glob
import pandas as pd

files=glob.glob('csv_folder/csv_for_concat/*.xlsx')
print(files)
 
# プログラム3｜変数listを空リストで設定
list = []
 
# プログラム4｜プログラム2で取得したエクセルを一つずつpandasデータとして取得
for file in files:
    list.append(pd.read_excel(file))
 
# プログラム5｜listに格納されたエクセルファイルをpandasとして結合
df = pd.concat(list)
 
# プログラム6｜エクセルファイルを書き出す
df.to_excel('total.xlsx', index=False)
```
